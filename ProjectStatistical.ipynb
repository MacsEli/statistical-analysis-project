```python
# Eliminamos el NaN encontrado
dataSet = dataSet.dropna(subset=['Electrical'])

# Los demás NaN los convertimos a 0
dataSet = dataSet.fillna(0)

# Ponemos el Id como indice para que no nos de problemas mas adelante
dataSet = dataSet.set_index('Id',drop=True)

# Obtenemos una columna binaria que nos indique si la casa ah sido remodelada
dataSet['Remodelacion'] = dataSet.YearRemodAdd == dataSet.YearBuilt
dataSet['Remodelacion'] = dataSet['Remodelacion'].astype('str')
dataSet['Remodelacion'] = dataSet['Remodelacion'].replace({'True':0,'False':1})

# Obtenemos una columna binaria que nos indique si tiene o no Aire Acondicionado
dataSet['CentralAir'] = dataSet['CentralAir'].replace({'N':0,'Y':1})

df_col_binarias = ['Remodelacion','CentralAir']

# Obtenemos una columna con los años de haber sido construida la casa
dataSet['AñosConstruidos'] = 2021 - dataSet['YearBuilt']
# Obtenemos una columna con los años de haber sido remodelada la casa
dataSet['AñosRemodelacion'] = 2021 - dataSet['YearRemodAdd']
dataSet['AñosRemodelacion'] = dataSet['AñosRemodelacion'].where(dataSet['Remodelacion']==1,0)
dataSet['AñosRemodelacion']

# Obtenemos en una columna los años de construccion del garaje
dataSet['AñosConstGar']= 2021 - dataSet['GarageYrBlt']
dataSet['AñosConstGar'] = dataSet['AñosConstGar'].replace({2021:0})

# Unificamos datos
# Sumaremos el numero total de baños completos
dataSet['TotBaños'] = dataSet['BsmtFullBath']+dataSet['FullBath']

# Sumaremos el numero total de medios baños
dataSet['TotMedBaño']=dataSet['BsmtHalfBath']+dataSet['HalfBath']

# Sumaremos el total de pies cuadrados de terminacion del sotano
dataSet['TotalBsmFinisSF'] = dataSet['BsmtFinSF1']+dataSet['BsmtFinSF2']

# Sumaremos el total de pies cuadrados de los pisos de la casa
dataSet['TotalFlrSF'] = dataSet['1stFlrSF']+dataSet['2ndFlrSF']

# Eliminamos las columnas sobre las cuales ya ejecutamos alguna transformacion
dataSet.drop(['BsmtFinSF1','BsmtFinSF2','1stFlrSF','2ndFlrSF','GarageYrBlt','BsmtFullBath','FullBath','BsmtHalfBath','HalfBath','MSSubClass','MoSold','YrSold','YearRemodAdd','YearBuilt'],axis=1,inplace=True)
```
